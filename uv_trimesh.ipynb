```python
import numpy as np
import trimesh
import matplotlib.pyplot as plt
```

```python
mesh_obj_path = 'demo_data/hrach/test2_0.obj'
```

```python
mesh = trimesh.load(mesh_obj_path)
```

```python
uv = mesh.vertices[:, :2]
mesh.visual.uv = uv
```

```python
uv_coords = mesh.visual.uv

fig, ax = plt.subplots()
ax.scatter(uv_coords[:, 0], uv_coords[:, 1], s=2)
ax.axis('off')
plt.savefig('graph.jpg', format='jpg', bbox_inches='tight', pad_inches=0)
```
